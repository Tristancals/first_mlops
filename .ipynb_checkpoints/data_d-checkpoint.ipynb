## Chargement des données

```python
d = lds(n_class=10)

print(d.keys())

print(d.data.shape)
print(d.data[1])

print(d.images.shape)
print(d.images[1])

print(d.target.shape)
print(d.target[1])

print(d.frame)

print(d.feature_names)

print(d.DESCR)
```

## Nettoyage et prétraitement

Comme vu dans le notebook précédent, écrivez le code permettant de nettoyer vos données et de les prétraiter

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(d.data, d.target, test_size=0.33) # utilise la méthode pour diviser le jeu de données en deux
```

Maintenant que le prétraitement est fini vous avez donc les splits `X_train`, `y_train`, `X_test` et `y_test` qui vous permettront de créer le meilleur modèle que possible. (si vous ne les avez pas, créez les)

Quelles sont les dimensions des splits?

## Création du modèle et sélection d'un modèle

Ici c'est le moment d'instancier un modèle et d'utiliser `fit` et `predict`!

```python
from sklearn.neighbors import KNeighborsClassifier
knn_clf = KNeighborsClassifier(n_neighbors=3) # instancie l'objet KNeighborsClassifier avec 3 voisins
knn_clf.fit(X_train,y_train) # utilise l'objet et sa méthode fit pour entrainer le modèle sur le jeu de données d'entrainement

```

Mais,... avez vous pensé au `GridSearch`? ;-) Quelle métrique allez vous évaluer?

```python
from sklearn.model_selection import GridSearchCV
parameters = {'n_neighbors':list(range(1,30))} # dictionnaire comprenant pour chaque paramètre (clé du dictionnaire) une liste de valeurs attendues.
neighbors_model = KNeighborsClassifier() # instancie un modèle KNN sans paramètre
gridsearch_models = GridSearchCV(neighbors_model, parameters, scoring='accuracy') # instancie un gridsearchCV avec le modèle et le jeu de paramètres

print(gridsearch_models)

gridsearch_models.fit(X_train,y_train) # utilise l'objet et sa méthode fit pour entrainer le modèle sur le jeu de données d'entrainement

print(gridsearch_models.cv_results_.keys())


print("\n prout!!", gridsearch_models.best_estimator_)

gridsearch_models_results_pd = pd.DataFrame(gridsearch_models.cv_results_['mean_test_score'], columns = ["accuracy"], index = list(range(1,30)))
accuracy_plot = gridsearch_models_results_pd.plot.bar();
accuracy_plot.axhline(y=gridsearch_models_results_pd.accuracy.max(), xmin=-1, xmax=31, color='r', linestyle='--', lw=2);
```

## Évaluation du modèle

L'heure d'utiliser le jeu de test a sonné! Évaluez la métrique sur le jeu de test

```python
from sklearn.metrics import confusion_matrix ,ConfusionMatrixDisplay,balanced_accuracy_score


y_pred = knn_clf.predict(X_test)
confusion_matrix_values = confusion_matrix(y_test,y_pred)
accuracy = balanced_accuracy_score(y_test,y_pred) # calcul de l'accuracy
print(accuracy)
# (Bonus) utilisez les outils de visualisation pour afficher la matrice de confusion
cm_display = ConfusionMatrixDisplay(confusion_matrix_values).plot()
```

## Export du modèle

Le titre de cette section parle de lui même :-)

```python
from sklearn import tree
# import du modèle d'arbre voulu
clf = tree.DecisionTreeClassifier(random_state=0)
tree.plot_tree(clf.fit(X_train,y_train))

y_pred = clf.predict(X_test) # utiliser la fonction predict pour faire prédire le modèle


#import de la métrique depuis scikit-learn.

accuracy = balanced_accuracy_score(y_test,y_pred) # calcul de l'accuracy
print(accuracy)
```
